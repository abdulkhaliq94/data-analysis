# Exercise 1

    1. Use variable “savings” to store the amount of bank savings (1000)
    2. Use variable “interest_multiplier” to store the multiplier 1.05
    3. Assign to variable “year1” the amount of money after 1 year.
    4. Print the result in the following sentence: “I started with $xxxx and now have \$xxxx.”
    5. Assign to variable “year7” the amount of money after 7 years.
    6. Print the result in a sentence similar to d).
    7. Find the types of all variables used in this script.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
savings = 1000
interest_multiplier = 1.05
year1 = (savings * 1.05) / 100 + savings
print(f"I started with {savings} and now have {year1}")
year7 = (savings * 1.05) / 100 * 7 + savings
print(f"I started with {savings} and now after 7 years have {year7}")
print(type(savings), type(interest_multiplier), type(year1), type(year7))
```

# Exercise 3

Plot the graph of sin(x) and cos(x) over interval [0, 3pi]. Add title, legend, and axis labels to the graph.

```python
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
x = np.arange(0, 3 * np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.title('Graph of sin(x) and cos(x)')
plt.legend(['y = sin(x)', 'y = cos(x)'], loc = 'upper left')
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```
