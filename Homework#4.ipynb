```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies_data = pd.read_csv("../Data/ml-latest-small/movies.csv", delimiter=",")
movies_data.head()
```

```python
ratings_data = pd.read_csv("../Data/ml-latest-small/ratings.csv", delimiter=",")
ratings_data.head()
```

### 1. Create column "AvgRating50" in the movies dataframe. For each movie, if there are less than 50 ratings, set its AvgRating50 value to -1.0; otherwise, set the value to the average rating.

```python
def ratings50(index):
    movie = movies_data.loc[index, "movieId"]
    movies_filter = ratings_data["movieId"] == movie
    ratings_filter = ratings_data[movies_filter]
    
    if(len(ratings_filter) >= 50):
        return ratings_filter["rating"].mean()
    else:
        return -1.0

_index = movies_data.index
movies_data["AvgRating50"] = pd.Series(_index).apply(ratings50)
movies_data.head()
```

### 2. Print the complete list of genres. (Use `split()` method to split a string containing multiple genres)

```python
genre_list=[]
def list_genres(index):
    genres = movies_data.loc[index, "genres"].split("|")
    
    for genre in genres:
        genre_list.append(genre)
        
pd.Series(_index).apply(list_genres)
genre_list = list(set(genre_list))

print("\n ".join(genre_list))
```

### 3. Which genre has the largest number of movies?

```python
genre_index = movies_data["genres"]
counts = []

for genre in genre_list:
    counter = 0
    
    for i in np.arange(0, len(genre_index)):
        if(genre in genre_index[i]):
            counter = counter + 1
            
    counts.append(counter)
print(counts)
```

```python
genre_list = pd.Series(genre_list)
counts = pd.Series(counts)

fil = {"genre": genre_list, "count": counts}
sub_dfr = pd.DataFrame(fil)

maximum = sub_dfr["count"] == sub_dfr["count"].max()
max_value = sub_dfr[maximum]
max_value
```

#### Observations:
- Drame has the largest number of movies

## 4. For each genre, find the ID of the movie with highest AvgRating50 value.

```python
max_index = []

genre_fields = movies_data["genres"]
avg_ratings50 = movies_data["AvgRating50"]

for genre in genre_list:

    max_movie = 0.0
    for i in np.arange(0,len(genre_fields)):
            
        if((genre in genre_fields[i]) & (avg_ratings50[i] > max_movie)):
            max_movie = i
    
    max_index.append(max_movie)
```

```python
# Max movie ids and sub dataframe to find ID of the movie with highest AvgRating50
max_movieIds = []
movieIds = movies_data["movieId"]

for index in max_index:
    max_movieIds.append(movieIds[index])
    


max_movieIds = pd.Series(max_movieIds)
sub_dfr["highest_rated_id"] = max_movieIds

sub_dfr
```

## 5. For each genre, find the name of the movie with the highest AvgRating50 value.

```python
names = []

for movie_id in max_movieIds:
    
    movie_idx = movies_data["movieId"] == movie_id
    movie_name = movies_data[movie_idx]["title"].values[0]
    names.append(movie_name)
    
names
```

```python
names = pd.Series(names)
sub_dfr["highest_rated_name"] = names

sub_dfr
```
